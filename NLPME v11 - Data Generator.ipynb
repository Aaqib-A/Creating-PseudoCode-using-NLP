```python
#function to create variables

def generate_variable(dataFill, append_extra=False):
    return_list = []
    s_included = False
    
    for each_dataFill in dataFill:
        if (s_included==False):
                for i in range(0,3):
                    return_list.append(each_dataFill+str(i))
                    return_list.append(each_dataFill+"_"+str(i))
                    s_included = True
                '''
                for i in range(10,13):
                    return_list.append(each_dataFill+str(i))
                    return_list.append(each_dataFill+"_"+str(i))
                    s_included = True
                '''
                
        else:
            s_included = False
            
    if (append_extra==True):        
        return_list.append("i")
        return_list.append("j")
        return_list.append("k")
        #return_list.append("a")
        return_list.append("b")
        return_list.append("c")
        
    return (return_list)
            
#print (variables_list)
#print(generate_variable(dataType))
#print(generate_variable(dataType_char))
#print(generate_variable(dataType_list, True))
#print(generate_variable(["var"]))

variables_list = generate_variable(dataType, True)
#print (variables_list)
```

```python
#function to generate function names

def generate_func_names(func_prog_list, operators_list):
    func_name_list = []
    for each_func in func_prog_list:
        for each_oper in operators_list:
            
            if each_oper == "+":
                each_oper = "add"
            elif each_oper == "-":
                each_oper = "minus"
            elif each_oper == "*":
                each_oper = "mul"
            elif each_oper == "/":
                each_oper = "div"
            elif each_oper == "%":
                each_oper = "mod"
            
            func_name = each_func +"_"+ each_oper
            func_name_list.append(func_name)
            
            func_name =  each_oper +"_"+ each_func
            func_name_list.append(func_name)
            
    func_name_list = list( dict.fromkeys(func_name_list) )   #Remove duplicates     
    return func_name_list

func_name = generate_func_names(func_prog, operators)

#print (variables_list)
#len(func_name)
```

```python
# Remove random lines from the file
#3,81,123 is a lot of lines to train
#Update 4,13,784
#Update 9,59,127 - 3 Mar 2021

import random

def reduce_data(original_data,div_factor):
    index_to_delete = []
    loop_size = len(original_data)
    current_iteration = 0
    while( (current_iteration + div_factor) <= loop_size ):
        rand_num = random.randint(1,div_factor)
        index_pos = current_iteration + rand_num    
        index_to_delete.append(index_pos)    
        current_iteration += div_factor
    #print("index_to_delete:\n", index_to_delete)
    #final_list = [i for j, i in enumerate(original_data) if j in index_to_delete]
    final_list = [original_data[i] for i in index_to_delete]
    return (final_list)

final_list = reduce_data(training_data, 10)
final_list += reduce_data(training_data2, 3)

print(len(final_list))
```

```python
#Main Program v0.1 Started:Jan-28-2021
#This Program is for Analysing files and its different Formats

#Reading the Data from input file
fileData=[]
with open("Input_file.txt", "r") as file:
    for line in file:
        #x = re.sub("=", " = ", line)
        #x = re.sub("  "," ", x)   
        fileData.append(line)

#for eachLine in fileData:
#    print (eachLine, end="")

import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
for eachLine in fileData:
    x = re.sub("=", " = ", eachLine)
    x = re.sub("  "," ", x)   
    nlp_eachLine = nlp(x)    
    
    #nlp_eachLine = nlp(eachLine)
    for token in nlp_eachLine:
        # Get the token text, part-of-speech tag and dependency label
        token_text = token.text
        token_pos = token.pos_
        token_dep = token.dep_
        print(f"{token_text:<12}{token_pos:<10}{token_dep:<10}")
```

```python
for eachLine in fileData:
    nlp_eachLine = nlp(eachLine)
    for ent in nlp_eachLine.ents:
        print(ent.text, ent.label_)
```

```python
#Checking Lemmatized Words
from spacy.lemmatizer import Lemmatizer
from spacy.lookups import Lookups
lookups = Lookups()
lookups.add_table("lemma_rules", {"noun": [["s", ""]]})
lemmatizer = Lemmatizer(lookups)
#lemmas = lemmatizer("ducks", "NOUN")
#print(lemmas)

for eachLine in fileData:
    x = re.sub("=", " = ", eachLine)
    x = re.sub("  "," ", x)   
    nlp_eachLine = nlp(x) 
    
    #nlp_eachLine = nlp(eachLine)
    for token in nlp_eachLine:
        # Get the token text, part-of-speech tag and dependency label
        token_text = token.text
        token_pos = token.pos_
        lemmas = lemmatizer(token_text, token_pos)
        print(lemmas, end="")
    print("")
```

```python
#Testing Creating Data for training

#matcher = Matcher(nlp.vocab)
#pattern1 = [{"LOWER": "iphone"}, {"LOWER": "x"}]
#pattern2 = [{"LOWER": "iphone"}, {"IS_DIGIT": True}]
#matcher.add("GADGET", None, pattern1, pattern2)

fileData=[]
with open("Input_file.txt", "r") as file:
    for line in file:
        x = re.sub("=", " = ", line)
        x = re.sub("  "," ", x)   
        fileData.append(x)

from spacy.matcher import Matcher
#from spacy.lang.en import English
import re

#nlp = English()

matcher = Matcher(nlp.vocab)
pattern1 = [{"IS_ALPHA": True}, {"ORTH": "="}, {"IS_DIGIT": True}]
pattern2 = [{"IS_ALPHA": True}, {"LOWER": "to"}, {"IS_DIGIT": True}]
matcher.add("INITIALIZATION", None, pattern1, pattern2)
TRAINING_DATA = []

# Create a Doc object for each text in TEXTS
for doc in nlp.pipe(fileData):
    #x = re.sub("=", " = ", doc)
    #doc = re.sub("  "," ", x)
    
    # Match on the doc and create a list of matched spans
    spans = [doc[start:end] for match_id, start, end in matcher(doc)]
    # Get (start character, end character, label) tuples of matches
    entities = [(span.start_char, span.end_char, "INITIALIZATION") for span in spans]
    
    #entities = [(span.start_char, span.end_char, nlp.vocab.strings[span.match_id]) for span in spans]
    #entities = [(span.start_char, span.end_char, span) for span in spans]
    for eachWord in entities:
        print("Entities: {}".format(eachWord))
    
    # Format the matches as a (doc.text, entities) tuple
    training_example = (doc.text, {"entities": entities})
    # Append the example to the training data
    TRAINING_DATA.append(training_example)

print(*TRAINING_DATA, sep="\n")

```

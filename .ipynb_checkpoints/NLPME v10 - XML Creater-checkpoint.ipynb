```python
import spacy
from spacy import displacy

nlp = spacy.load("./NLP_Object")

def entity_identify(statement):
    doc = nlp(statement) 
    
    ent_text=[]
    ent_label=[]
    for ent in doc.ents:
        #print (ent.text, ent.start_char, ent.end_char, ent.label_)
        ent_text.append(ent.text)
        ent_label.append(ent.label_)
        
    return (ent_text,ent_label)
```

```python
def xml_vari_creation(var_name,default_scope, default_datatype, var_value):
    root = minidom.Document() 
    productChild = root.createElement('variable')
    if var_name != "NA":
        productChild.setAttribute('name', var_name) 
    else:
        productChild.setAttribute('name', "default") 
    
    if default_scope != "NA":
        productChild.setAttribute('scope', default_scope) 
    #else:
        #productChild.setAttribute('scope', "default_scope") 
    
    if default_datatype != "NA":
        productChild.setAttribute('datatype', default_datatype) 
    #else:
        #productChild.setAttribute('datatype', "default_datatype") 
        
    if var_value != "NA":
        productChild.setAttribute('value', var_value) 
    #else:
        #productChild.setAttribute('value', "default_value") 
    return productChild
    #variables_list.appendChild(productChild)
```

```python
#Testing XML Function

from xml.dom import minidom 
import os 

def nlpToXML(statement):
    
    root = minidom.Document() 
    xml = root.createElement('root')  
    root.appendChild(xml) 
    #-------------------

    #statement = "create a function to add numbers a and b"
    
    ent_text, ent_label = entity_identify(statement)
    #entity_display(statement)
    #print(ent_text ," ", ent_label)

    #=== Basics ===#
    '''
    productChild = root.createElement('product') 
    productChild.setAttribute('name', 'Geeks for Geeks') 

    xml.appendChild(productChild)
    '''
    #=== Basics ===#

    #Statement - Start
    statementChild = root.createElement('statement') 
    statementChild.setAttribute('value', statement) 

    xml.appendChild(statementChild)
    #Statement - End
    
    #Prog - Start
    prog = root.createElement('program') #
    if "FUNC_PROG" in ent_label:
        index = ent_label.index("FUNC_PROG")
        #type_of_prog_type = ent_text[index]

        prog.setAttribute('type', ent_text[index]) #

    else:
        prog.setAttribute('type', "program") #

    prog.setAttribute('name', "test") #
    xml.appendChild(prog)  #
    #Prog - End

    #Variables - Start
    variables_list = root.createElement('variables') 
    prog.appendChild(variables_list) 

    var_name = "NA"
    default_scope = "NA"
    default_datatype = "NA"    
    var_value = "NA"
    variable_flag = [0, 0, 0, 0]
    for index in range(len(ent_text)):
        #==
        if ent_label[index]== "SEPERATOR" and 1 in variable_flag:
            productChild = xml_vari_creation(var_name,default_scope, default_datatype, var_value)
            variable_flag = [0, 0, 0, 0]
            var_name = "NA"
            var_value = "NA"
            variables_list.appendChild(productChild)
        #==
        if ent_label[index]== "VARI":
            variable_flag[0]+=1     
            if variable_flag[0]>=2:
                productChild = xml_vari_creation(var_name,default_scope, default_datatype, var_value)
                variable_flag = [1, 0, 0, 0]
                var_name = "NA"
                var_value = "NA"
                variables_list.appendChild(productChild)
            var_name = ent_text[index]
        #==    
        if ent_label[index] == "SCOPE": 
            variable_flag[1]+=1 
            if variable_flag[1]>=2:
                productChild = xml_vari_creation(var_name,default_scope, default_datatype, var_value)
                variable_flag = [0, 1, 0, 0]
                var_name = "NA"
                var_value = "NA"
                variables_list.appendChild(productChild)
            default_scope = ent_text[index] 
        #==    
        if ent_label[index] == "DATATYPE": 
            variable_flag[2]+=1
            if variable_flag[2]>=2:
                productChild = xml_vari_creation(var_name,default_scope, default_datatype, var_value)
                variable_flag = [0, 0, 1, 0]
                var_name = "NA"
                var_value = "NA"
                variables_list.appendChild(productChild)
            default_datatype = ent_text[index]
        #==    
        if ent_label[index] == "VALUE":
            variable_flag[3]+=1
            if variable_flag[3]>=2:
                productChild = xml_vari_creation(var_name,default_scope, default_datatype, var_value)
                variable_flag = [0, 0, 0, 1]
                var_name = "NA"
                var_value = "NA"
                variables_list.appendChild(productChild)
            var_value = ent_text[index]
        #==
    if var_name != "NA":
        productChild = xml_vari_creation(var_name,default_scope, default_datatype, var_value)
        variable_flag = [0, 0, 0, 0]
        var_name = "NA"
        var_value = "NA"
        variables_list.appendChild(productChild)
    
    var_name = "NA"
    default_scope = "NA"
    default_datatype = "NA"    
    var_value = "NA"             
    #Variables - End

    for index in range(len(ent_text)):
        if ent_label[index]== "COMMAND":
            productChild = root.createElement('command') 
            productChild.setAttribute('name', ent_text[index]) 
            prog.appendChild(productChild) 

        if ent_label[index]== "OPERATOR":
            productChild = root.createElement('operation') 
            productChild.setAttribute('name', ent_text[index]) 
            prog.appendChild(productChild) 
             
    #-------------------
    xml_str = root.toprettyxml(indent ="\t")  
    #print (xml_str)
    return(xml_str)
    '''
    save_path_file = "xml_test.xml"  
    with open(save_path_file, "w") as f: 
        f.write(xml_str)  
    '''
```
